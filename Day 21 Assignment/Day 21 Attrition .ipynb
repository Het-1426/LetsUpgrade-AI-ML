# will convert the object dtypes to int dtypes

```python
from sklearn import preprocessing
```

```python
label_enoder = preprocessing.LabelEncoder()
data['Attrition'] = label_enoder.fit_transform(data['Attrition'])  #Converted Attrition to int type from object type
data['Department'] = label_enoder.fit_transform(data['Department']) #converted department to int type as(1,2,3)
data['Gender'] = label_enoder.fit_transform(data['Gender']) #converted gender to int type 
data['MaritalStatus'] = label_enoder.fit_transform(data['MaritalStatus']) #converted marital status to int type
data['NumCompaniesWorked'] = label_enoder.fit_transform(data['NumCompaniesWorked'])
data['TotalWorkingYears'] = label_enoder.fit_transform(data['TotalWorkingYears'])
```

```python
data.head(8)
```

# Now we will start the Logistic Regression Anylysis

## Assigning DV & IDV

```python
y  = data['Attrition']  #dependent variable
x = data[['Age', 'Department', 'DistanceFromHome', 'Education', 'Gender', 'JobLevel', 'MaritalStatus', 'MonthlyIncome', 'NumCompaniesWorked', 'PercentSalaryHike', 'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear', 'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']] #independent variable
```

```python
#Adding constant 
import statsmodels.api as sm
x1 = sm.add_constant(x)
```

```python
Logistic = sm.Logit(y,x1)
```

```python
Result = Logistic.fit()
```

```python
Result.summary()
```
