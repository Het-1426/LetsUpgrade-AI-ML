```python
import pandas as pd #imported pandas
```

```python
dataset = pd.read_excel("Bank_Personal_Loan_Modelling.xlsx", sheet_name=1)
dataset.head()                     #imported required dataset
```

```python
dataset.isna().sum()  #checking for null values
```

```python
dataset.drop(['ID', 'ZIP Code'], axis=1, inplace=True) #Droped the columns sir told to 
```

```python
dataset.head()
```

```python
y = dataset[['Personal Loan']]  #dependent variable defined
x = dataset[['Age', 'Experience', 'Income', 'Family', 'CCAvg', 'Education', 'Mortgage', 'Securities Account', 'CD Account', 'Online', 'CreditCard']]
```

```python
y.head(2)
```

```python
x.head(2)
```

```python
import statsmodels.api as sm 
```

```python
x1 = sm.add_constant(x)
```

```python
Logistic = sm.Logit(y,x1)
```

```python
Result = Logistic.fit()
```

```python
Result.summary()
```
