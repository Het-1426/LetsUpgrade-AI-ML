```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn import neighbors
```

```python
da = pd.read_csv('train.csv')
da.head()
```

```python
da.isnull().sum()
```

```python
da.dropna().head()
```

```python
da.duplicated().sum()
```

```python
le = preprocessing.LabelEncoder()
```

```python
da['Sex'] = le.fit_transform(da['Sex'])
da['Embarked'] = le.fit_transform(da['Embarked'])
```

```python
y = da['Pclass']
x = da[['Survived', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
def k_value (a):
    knn = neighbors.KNeighborsClassifier(n_neighbors=a)
    knn.fit(x_train,y_train)
    y_pred = knn.predict(x_test)
    print(f'Accuracy when k-value is {a} = {accuracy_score(y_test,y_pred)}')
    print(f'Confusion Matrix when k-value is {a} = \n{confusion_matrix(y_test,y_pred)}')
```

```python
lis = list(range(1,268))
```

```python
for i in lis:
    k_value(i)
```
